```python
import json

input_path = "../results/determine-url-deps/browser-logs.json.d/01.json"

with open(input_path, "r") as jsonfile:
    lines = jsonfile.readlines()

json01 = json.loads(lines[5])

for k in json01.keys():
    print(k)
print("---------------------\n")

print(json01['url'])
print(json01['final_url'])
print("---------------------")
for log_message in json01["http_trace"]:
    #print(log_message["message"]["message"]["params"])
    message = log_message["message"]["message"]
    if message["method"] != "Network.requestWillBeSent":
        continue
    request = message["params"]["request"]
    print(request["url"])
    print("")
    initiator = message["params"]["initiator"]
    print(initiator)
    print("--")
```
