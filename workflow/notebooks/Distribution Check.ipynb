```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(0)

def plot(k, lower, upper):
    fig, axes = plt.subplots(1, 1, figsize=(3.5, 2), dpi=150)
    data = np.concatenate(
        (-rng.gamma(k/2, scale=50, size=10000) + upper,
        rng.gamma(k/2, scale=50, size=10000)+ lower)
    )
    sns.histplot(data, ax=axes, bins=100)
    axes.set_xticks(np.arange(lower, upper, 150))
    axes.set_xlim(lower, upper)
    
plot(3, lower=-700, upper=160)
```

```python
rng = np.random.default_rng(0)

def plot(k, lower, upper):
    fig, axes = plt.subplots(1, 1, figsize=(3.5, 2), dpi=150)
    data = (upper-lower)*rng.beta(0.5, 0.5, size=20_000) + lower
    sns.histplot(data, ax=axes, bins=100)
    axes.set_xticks(np.arange(lower, upper, 150))
    axes.set_xlim(lower, upper)
    
plot(2, lower=-700, upper=160)
```

```python
rng = np.random.default_rng(0)

def plot(k, lower, upper):
    fig, axes = plt.subplots(1, 1, figsize=(3.5, 2), dpi=150)
    val = upper - lower
    data = rng.normal(val, 100, size=10_000) % val + lower
    sns.histplot(data, ax=axes, bins=100)
    axes.set_xticks(np.arange(lower, upper+1, 150))
    # axes.set_xlim(lower, upper)
    
plot(2, lower=-750, upper=150)
```
