```python
import h5py
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with h5py.File("../../results-2/ml-eval-conn/tamaraw/dataset.h5", mode="r") as h5in:
    sizes = h5in["sizes"][:]
```

```python
dist = [len(x) for x in sizes]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, dpi=150, figsize=(10, 3.5))
sns.histplot(dist, ax=ax1, binwidth=1000, stat="probability", binrange=(0, 40000))
sns.histplot(dist, ax=ax2, binwidth=1000, stat="density", cumulative=True, binrange=(0, 40000))
# ax.set_xlim(0, 50000)
```
